```python
from qiskit import *
%matplotlib inline
```

```python
circuit = QuantumCircuit(2,2)
circuit.h(0)
circuit.cx(0,1)
circuit.draw(output='mpl')
```

```python
circuit = QuantumCircuit(2,2)
circuit.x(0)
circuit.h(0)
circuit.cx(0,1)
circuit.draw(output='mpl')
```

```python
circuit = QuantumCircuit(2,2)
circuit.h(0)
circuit.x(1)
circuit.cx(0,1)
circuit.draw(output='mpl')
```

```python
circuit = QuantumCircuit(2,2)
circuit.h(0)
circuit.x(1)
circuit.z(0)
circuit.z(1)
circuit.cx(0,1)
circuit.draw(output='mpl')
```

```python
circuit = QuantumCircuit(2,2)
circuit.h(0)
circuit.cx(0,1)
circuit.barrier()
circuit.cx(0,1)
circuit.h(0)
circuit.draw(output='mpl')
```
